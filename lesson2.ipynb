#### Default Argument Values
It is also possible to define functions with a <code style="background-color:#d9ffe8;">variable number of arguments</code>. 

You can specify a <code style="background-color:#d9ffe8;">default value</code> for one or more arguments. This creates a function that can be called with fewer arguments than it is defined to allow.

```python
def str_sum(p1, p2, p3=10):
    
    print('p1 =', p1, end=', ')
    print('p2 =', p2, end=', ')
    print('p3 =', p3, end=', ')
    
    result = "{} + {} + {} = {}".format(p1, p2, p3, p1+p2+p3)
    return result
```

```python
p1, p2, p3 = (1, 5, 15)
```

```python
str_sum(p1, p2, p3)
```

```python
str_sum(p1, p2)
```

The default values are evaluated at the point of function definition in the defining scope, so that

```python
i = 5

def f(arg=i):
    print(arg)

i = 6
f()
```

<b>Important warning</b>: The default value is evaluated only once. This makes a difference when the default is a mutable object such as a list, dictionary, or instances of most classes. For example, the following function accumulates the arguments passed to it on subsequent calls:

```python
def f(a, L=[]):
    L.append(a)
    return L

print(f(1))
```

```python
print(f(3))
```

if you don’t want the default to be shared between subsequent calls, you can write the function like this instead:

```python
def f(a, L=None):
    if L is None:
        L = []
    L.append(a)
    return L
```

```python
print(f(1))
```

```python
print(f(2))
print(f(3))
```

In a function call, keyword arguments must follow <code style="background-color:#d9ffe8;">positional arguments</code>. 

All the keyword arguments passed must match one of the arguments accepted by the function, and their order is not important. 

```python
def str_sum(p1, p2, p3=10):
    
    print('p1 =', p1, end=', ')
    print('p2 =', p2, end=', ')
    print('p3 =', p3, end=', ')
    
    result = "{} + {} + {} = {}".format(p1, p2, p3, p1+p2+p3)
    return result
```

<code style="background-color:#d9ffe8;">List Comprehensions</code> 

<code style="background-color:#d9ffe8;">List Comprehensions</code> provide a concise way to create sequences. Common applications are to make new lists where each element is the result of some operations applied to each member of another sequence or iterable, or to create a subsequence of those elements that satisfy a certain condition.

For example, assume we want to create a <code style="background-color:powderblue;">list</code> of squares, like:

```python
squares = []
for x in range(10):
    squares.append(x**2)

squares
```

Note that this creates (or overwrites) a variable named x that still exists after the loop completes. We can calculate the list of squares without any side effects using

```python
squares = [x**2 for x in range(0, 10)]
```

```python
squares
```

which is more concise and readable.

A list comprehension consists of brackets containing an expression followed by a for clause, then zero or more for or if clauses. The result will be a new list resulting from evaluating the expression in the context of the for and if clauses which follow it.

```python
# comprehension
a_dict = {i:str(i*2) for i in range(10)}
print(a_dict)
```

```python
a_dict = {}
for i in range(10):
    a_dict[i] = str(i*2)
    
print(a_dict)
```

```python
def is_prime(n):
    
    if n == 0 or n == 1:
        return False
    
    elif n > 1:
        
        for i in range(2, n):
            
            if n % i == 0:
                return False
    
    return True
```

```python
n = 7
is_prime(n)
```

```python
def test_while(n):
    '''Returns next prime'''
    
    while not is_prime(n):
        n+=1 #n = n+1
    
    return n
```

```python
n = 24
test_while(n)
```

```python
def test_break(n):
    
    while True:
        if is_prime(n):
            break
        else:
            n+=1
    
    return n
```

#### Continue
With the <code style="background-color:powderblue;">continue</code> statement we can stop the current iteration of the loop, and continue with the next

```python
[1, 2, 3, 4, 5, 6]
```

```python
def test_continue(l):
    '''Print even numbers'''
    
    for i in l:
        if i % 2 == 1:
            continue
        
        print(i, end=', ')
    
```

```python
test_continue(list(range(1, 50)))
```

#### Try - Except
The <code style="background-color:powderblue;">try</code> block lets you test a block of code for errors.

The <code style="background-color:powderblue;">except</code> block lets you handle the error.

```python
def test_try_except(safe=False):
    if safe:
        
        try:
            a = 3/0
        except:
            print('Error: division by zero')
    
    else:
        a = 3/0
    
    print('...')
    print('Do something')
```

```python
test_try_except(safe=True)
```

```python
try:
    raise IndentError
    pass

except ValueError:
    # handle ValueError exception
    print('ZeroDivisionError')

except (TypeError, ZeroDivisionError):
    # handle multiple exceptions
    # TypeError or ZeroDivisionError
    print('TypeError or ZeroDivisionError')

except:
    # handle all other exceptions
    print('Other')
```
